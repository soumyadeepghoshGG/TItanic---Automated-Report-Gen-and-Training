# Sweetviz Overview


[Sweetviz](https://pypi.org/project/sweetviz/) is a Python library for visualizing and comparing datasets with ease. It provides insightful visualizations to help data scientists and analysts understand the structure and characteristics of their data quickly. Sweetviz generates both visual and numerical statistics, making it a powerful tool for exploratory data analysis (EDA).

## Key Features

- **Automated EDA**: Sweetviz automates the process of exploratory data analysis, saving time and effort in understanding dataset properties.

- **Detailed Insights**: It generates detailed visualizations and statistics for both individual features and feature interactions.

- **Comparative Analysis**: Sweetviz facilitates the comparison of two datasets, such as training and testing sets, making it handy for model evaluation.

- **HTML Reports**: The results are presented in interactive HTML reports, making it easy to share and revisit the analysis.

#### 5. **Extensive Library Support:**
PyCaret supports a wide range of machine learning algorithms, making it versatile for different types of datasets and problems. It integrates with popular libraries like scikit-learn, XGBoost, LightGBM, spaCy, and more.

### Since Logistic Regression is the best model I will do <u>hyperparameter tuning</u> of the model.

```python
lr = create_model('lr', solver='liblinear')
```

**Solver Problem:** When creating a logistic regression model without specifying a solver, I encountered an underflow error during matrix multiplication. Logistic regression models aim to find the optimal set of coefficients by minimizing the logistic loss function. The solver is the algorithm used to perform this optimization. It determines how the model iteratively adjusts the coefficients to reach the optimal values. By explicitly setting the solver to 'liblinear', the issue was resolved, preventing numerical instability and allowing the logistic regression model to be successfully created. Here are examples of logistic regression solvers:
- *'newton-cg':* Newton's Conjugate Gradient. Suitable for small to medium-sized datasets.
- *'lbfgs':* Limited-memory Broyden-Fletcher-Goldfarb-Shanno. Generally works well for medium to large datasets. **DEFAULT**
- *'liblinear':* Library for Large Linear Classification. A good choice for small to medium-sized datasets. It uses a coordinate descent algorithm and is the default solver for scikit-learn's logistic regression.
- *'sag':* Stochastic Average Gradient. An optimization algorithm that is a variant of stochastic gradient descent. Particularly useful for large datasets.
- *'saga':* An extension of 'sag' that also supports L1 regularization. It's suitable for large datasets and can handle both L1 and L2 penalties.
- *'sgd':* Stochastic Gradient Descent. It's suitable for large datasets and online learning.

**Options of <u>plot</u> parameter in plot_model method**
- auc': Area Under the Curve (AUC) Plot - Receiver Operating Characteristic (ROC) curve.
- 'confusion_matrix': Confusion Matrix Plot - Visual representation of model performance.
- 'threshold': Threshold Plot - Visual representation of the effect of different probability thresholds.
- 'pr': Precision-Recall Curve - Trade-off between precision and recall.
- 'error': Class Prediction Error Plot - Shows the distribution of the predicted class against the actual class.
- 'class_report': Classification Report - Displaying precision, recall, F1-score, and support for each class.
- 'boundary': Decision Boundary Plot - Visual representation of the decision boundary.
- 'manifold': Manifold Learning Plot - Visualization of high-dimensional data in 2D or 3D.
- 'rfe': Recursive Feature Elimination Plot - Feature importance based on recursive feature elimination.
- 'learning': Learning Curve - Training and validation performance over iterations.
